```python
df = pd.read_csv('../input/titanic/train.csv')
df
```

```python
df=df.fillna(0)
df
```

```python
df=df.drop(['Name','Ticket'], axis = 1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
le_fare = LabelEncoder()
```

```python
df['sex_n'] = le_sex.fit_transform(df['Sex'])
df['fare_n'] = le_fare.fit_transform(df['Fare'])


df.head()
```

```python
df=df.drop(['Cabin','Embarked'], axis = 1)
df.head()
```

```python
df=df.drop(['Sex','Fare'], axis = 1)
df.head()
```

```python
y=pd.read_csv('../input/titanic/train.csv')

y.head()
```

```python
y=y.drop(['Sex','Fare','PassengerId','Pclass','Name','Age','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 1)
y.head()
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(df,y)
```

```python
model.score(df,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(['Survived'],axis=1),y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=180)
model1.fit(x_train,y_train)
```

```python
predictions = model1.score(x_test,y_test)
predictions
```

```python
y_predicted = model1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```
